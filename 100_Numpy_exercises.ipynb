#### 98. Considering a path described by two vectors (X,Y), how to sample it using equidistant samples (★★★)?

````python
t=np.linspace(0, 1, 10)**2
x=t
y=t**2

dx = np.diff(x)
dy = np.diff(y)
distances = np.sqrt(dx**2 + dy**2)

accumulated_dist = np.concatenate(([0], np.cumsum(distances)))

num_samples = 50
equidistant_dist = np.linspace(0, accumulated_dist[-1], num_samples)

x_new = np.interp(equidistant_dist, accumulated_dist, x)
y_new = np.interp(equidistant_dist, accumulated_dist, y)
````
